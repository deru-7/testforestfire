# Data Cleaning

```python
# missing values
df[df.isnull().any(axis=1)]
```

###  The dataset is converted into two sets based on Region from 122th index, we can make a new column based on the Region

#### 1. "Bejaia Region Dataset"

#### 2. "Sidi-Bel Abbes Region Dataset"

##### Add new column with region

```python
df.loc[:122,"Region"]=0
df.loc[122:,"Region"]=1
```

```python
df.info()
```

```python
df['Region']=df['Region'].astype(int)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
# removing the null values
df=df.dropna().reset_index(drop=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
df=df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
df.columns=df.columns.str.strip()
df.columns
```

```python
df.info()
```

### Change the requried columns as integer data type

```python
df.columns
```

```python
df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']]=df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
```

```python
df.info()
```

### Changing the other columns to float data type

```python
objects=[features for features in df.columns if df[features].dtype=='O']
```

```python
objects
```

```python
for i in objects:
    if i!='Classes':
        df[i]=df[i].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.head()
```

```python
# let save the cleaned dataset
df.to_csv('Algerian_forest_fires_cleaned_dataset.csv',index=False)
```

```python
# Percentage for Pie Chart
percentage=df_copy['Classes'].value_counts(normalize=True)*100
```

```python
df['Classes']=np.where(df['Classes'].str.contains('not fire'),'not fire','fire')
```

```python
# monthly Fire Analysis
plt.plot(figsize=(13,6))
sns.set_style('whitegrid')
sns.countplot(x='month',hue='Classes',data=df)
plt.ylabel('Number of Fries',weight='bold')
plt.xlabel('Months',weight='bold')
plt.title('Fire Analysis of all Regions',weight='bold')
plt.show()
```

```python
dftemp_1=df.loc[df['Region']==1]
plt.plot(figsize=(13,6))
sns.set_style('whitegrid')
sns.countplot(x='month',hue='Classes',data=dftemp_1)
plt.ylabel('Number of Fries',weight='bold')
plt.xlabel('Months',weight='bold')
plt.title('Fire Analysis of Sid-bel Region',weight='bold')
plt.show()
```

```python
dftemp_0=df.loc[df['Region']==0]
plt.plot(figsize=(13,6))
sns.set_style('whitegrid')
sns.countplot(x='month',hue='Classes',data=dftemp_0)
plt.ylabel('Number of Fries',weight='bold')
plt.xlabel('Months',weight='bold')
plt.title('Fire Analysis of Brjaia Region',weight='bold')
plt.show()
```

##### its obseverd that August and September had the most number of forest fires for both regions. And from the above plot of months, we can understand few things
